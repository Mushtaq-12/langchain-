## LLM Call

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

# 1. Load the key from your .env file
load_dotenv()

# 2. Initialize the client
client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY")) # It automatically looks for "OPENAI_API_KEY" in your environment

try:
    # 3. Make a simple request
    response = client.chat.completions.create(
        model="gpt-4o-mini",  # Best for learning (cheapest & fastest)
        messages=[
            {"role": "system", "content": "who have won the ipl in 2012 and 2014"},
            # {"role": "user", "content": "Who is current indian odi cricket caption "}
        ],
    )

    # 4. Print the result
    print("\nAI Response:")
    print(response.choices[0].message.content)

except Exception as e:
    print(f"\nOops! Something went wrong: {e}")
```
