### Chain in Langchain

```python
from langchain_openai import ChatOpenAI
import os 
from dotenv import load_dotenv

load_dotenv()

if os.environ.get("OPENAI_API_KEY"):
    print("Key exists")
else:
    raise ValueError("Key not Found!!")

from langchain_core.messages import HumanMessage,SystemMessage,AIMessage
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm_openai=ChatOpenAI(model='gpt-5-mini',temperature=0)
```

### First chain

```python
#Task-1 [prompt]
prompt_template=ChatPromptTemplate.from_messages([
    ("system","you are a helpful assistant"),
    ("human","{input}")
])
```

```python
#Task-2 llm
llm_openai=ChatOpenAI(model="gpt-5-mini",temperature=0)
```

```python
str_parser=StrOutputParser()
```

### Manual Invocation

```python
template=prompt_template.invoke({"input":"what is capital of france"})
res=llm_openai.invoke(template)
final_result=res.content
```

### Chain invocation

```python
chain=prompt_template | llm_openai |str_parser
chain.invoke({"input":"What is capital of france?"})
```

```python
#Another or Alternative way
from langchain_core.runnables import RunnableSequence

chain_1=RunnableSequence(prompt_template,llm_openai,str_parser)
chain_1.invoke("What is capital city of france?")
```
